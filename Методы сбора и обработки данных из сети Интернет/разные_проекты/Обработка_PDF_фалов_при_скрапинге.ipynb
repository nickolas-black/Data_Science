## Загружаем необходимые библиотеки

```
from pdfminer.pdfinterp import PDFResourceManager, process_pdf
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from io import StringIO
from io import open
from urllib.request import urlopen
```

## вспомогательная функция для обработки при скрапинге

```
def readPDF(pdfFile):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, laparams=laparams)

    process_pdf(rsrcmgr, device, pdfFile)
    device.close()

    content = retstr.getvalue()
    retstr.close()
    return content
```

## задаем страницу для скрапинга и преобразования

```
pdfFile = urlopen("http://repo.ssau.ru/bitstream/Uchebnye-posobiya/Analiz-dannyh-Elektronnyi-resurs-ucheb-posobie-po-programme-vyssh-obrazovaniya-59138/1/%D0%9F%D0%BE%D1%80%D1%83%D1%87%D0%B8%D0%BA%D0%BE%D0%B2%20%D0%9C.%D0%90.%20%D0%90%D0%BD%D0%B0%D0%BB%D0%B8%D0%B7%20%D0%B4%D0%B0%D0%BD%D0%BD%D1%8B%D1%85.pdf")
```

```
outputString = readPDF(pdfFile)
print(outputString)
pdfFile.close()
```
